```python
#1. Normal Distribution

import numpy as np
import matplotlib.pyplot as plt

# Set the seed for reproducibility
np.random.seed(1234)

# Generate random samples from a normal distribution
mean = 0  # Mean of the distribution
std_dev = 1  # Standard deviation of the distribution
num_samples = 10000  # Number of samples

samples = np.random.normal(mean, std_dev, num_samples)

# Create a histogram to visualize the distribution
num_bins = 50
clr = "#EFEFEF"

counts, edges, patches = plt.hist(samples, bins=num_bins, color=clr, density=True)
centers = 0.5 * (edges[:-1] + edges[1:])
plt.plot(centers, counts, 'k-', linewidth=2)

# Show the plot
plt.xlabel('Value')
plt.ylabel('Probability')
plt.title('Normal Distribution')
plt.show()
```

```python
#2.For the above distribution calculate each observation using the probability density
#function. A plot of these values would give us the tell-tale bell shape. The start and end
#range must be given by users.

import numpy as np
import matplotlib.pyplot as plt

mean = 50
std_dev = 5

start_range = 40
end_range = 60

x = np.linspace(start_range, end_range, 1000)
pdf = 1 / (std_dev * np.sqrt(2 * np.pi)) * np.exp(-(x - mean) ** 2 / (2 * std_dev ** 2))

plt.plot(x, pdf, color='red')
plt.xlabel('Value')
plt.ylabel('PDF')
plt.show()
```

```python
#3. Write a python program to display 20 numbers from the exponential distribution. In
#addition, for the observations between 50 and 70 do calculate the probability and
#cumulative probability for each and plot the result.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import expon

# Generate 20 random numbers from the exponential distribution
data = np.random.exponential(scale=1, size=20)
#range of interest
lower_bound = 50
upper_bound = 70

# observations that fall within the specified range
observations_between_50_and_70 = [obs for obs in data if lower_bound <= obs <= upper_bound]

# probability of an observation falling in the range
total_observations = len(data)
observations_in_range = len(observations_between_50_and_70)
probability_in_range = observations_in_range / total_observations

# cumulative probability for observations in the range
cumulative_probability = expon.cdf(upper_bound, scale=1) - expon.cdf(lower_bound, scale=1)

print(f"Generated data: {data}")
print(f"Observations between {lower_bound} and {upper_bound}: {observations_between_50_and_70}")
print(f"Probability of an observation in the range: {probability_in_range:.2f}")
print(f"Cumulative Probability in the range: {cumulative_probability:.2f}")

plt.hist(data, bins=10, density=True, color='blue', alpha=0.5, label='Exponential Data')
plt.axvline(lower_bound, color='red', linestyle='--', label=f'Lower Bound ({lower_bound})')
plt.axvline(upper_bound, color='green', linestyle='--', label=f'Upper Bound ({upper_bound})')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.legend()
plt.show()
```

```python
#4.  A survey found that 65% of all the financial consumers were very satisfied with their
#primary financial institutions. Suppose that 25 financial consumers are sampled and if
#survey results still hold true today, what is the probability that exactly 19 are very
#satisfied with their primary financial institutions?

from scipy.stats import binom

n = 25  # Total number of consumers
p = 0.65  # Probability of being very satisfied
k = 19  # Number of successes

probability = binom.pmf(k, n, p)
print(f"Probability: {probability}")
```

```python
#5.  Bank customers arrive randomly on weekday afternoons at an average of 3.2 customers
#per 4 minutes. What is the probability of having more than 7 customers in a 4-minute
#interval on a weekday afternoon? Accept the inputs from the user and plot the appropriate
#probability distribution functions.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Input from the user
average_customers = float(input("Enter the average number of customers per 4 minutes: "))
observed_customers = int(input("Enter the number of customers you want to calculate the probability for: "))

# Calculate the Poisson parameter (lambda) for the given average
lambda_ = (average_customers / 4) * 4  # Adjust the average to the 4-minute interval

# Calculate the probability of having more than 'observed_customers' customers
probability = 1 - poisson.cdf(observed_customers, lambda_)

# Plot the Poisson probability distribution function
x = np.arange(0, 20)  # Define a range of possible customer counts
pmf = poisson.pmf(x, lambda_)

plt.bar(x, pmf, color='blue')
plt.xlabel('Number of Customers')
plt.ylabel('Probability')
plt.title(f'Poisson Distribution ( = {lambda_})')
plt.show()

print(f"Probability of having more than {observed_customers} customers in a 4-minute interval: {probability:.4f}")
```
